```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random
import random
import sklearn
import pandas as pd
```

```python
from sklearn.datasets import load_iris

# Load the Iris dataset
iris = load_iris()

# Access the features (X) and target variable (y)
X = iris.data  # Features
y = iris.target  # Target variable

# Print the shape of the data
print("Data shape:", X.shape)

# Print the target names
print("Target names:", iris.target_names)
```

```python
x_train=iris.data
print(x_train)
print(x_train.shape)
```

```python
data_matrix = np.array(x_train)
print(data_matrix)
```

```python
mean = np.mean(x_train, axis=0)
x_train = x_train - mean
print(x_train)
```

```python
#size of data matrix=(150, 4)
#so covariance matrix size must be 4 x 4


# Calculate the covariance matrix
cov_matrix = np.cov(x_train, rowvar=False)

# Print the shape of the covariance matrix
print("Covariance matrix shape:", cov_matrix.shape)
print(cov_matrix)
```

```python
#visualize covariance matrix
plt.imshow(cov_matrix, cmap='Blues', interpolation='nearest')
extra_legend = ""
plt.text(0.9,1.03, extra_legend, ha='left', va='center', color='red', transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', edgecolor='red', boxstyle='round'))
plt.colorbar()
plt.title('Covariance Matrix')
plt.show()
```

```python
# Calculate the eigenvectors and eigenvalues
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

# Sort the eigenvalues and corresponding eigenvectors in descending order
idx = np.argsort(eigenvalues)[::-1]  # Reverse the order to sort in descending order
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]
```

```python
eigenvectors.shape
```

```python
print(eigenvectors)
```

```python
#proportion of variance of each eigenvalue
pov_list=[]
for i in range (0,4):
  pov=eigenvalues[i]/sum(eigenvalues)
  pov_list.append(pov)
```

```python
#print cumulative proportion of variance
#calculating the proporiton of
proportion_of_variance_list=[]
for k in range (1,5):
  selected_eigenvalues = eigenvalues[:k]
  proportion_of_variance=sum(selected_eigenvalues)/sum(eigenvalues)
  proportion_of_variance_list.append(proportion_of_variance)
```

```python
x = range(1, 5)
plt.plot(x, pov_list, color="blue", label='explained variance')
plt.plot(x, proportion_of_variance_list, color="red", label='cumulative explained variance')
plt.grid()
# Set the axis labels
plt.xlabel('eigenvalues')
plt.ylabel('Proportion of variance')
extra_legend = ""
plt.text(0.8,0.6, extra_legend, ha='left', va='center', color='red', transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', edgecolor='red', boxstyle='round'))
plt.title('proportion of variance vs. eigenvalues')
plt.legend()

# Display the plot
plt.show()
```

```python
print(eigenvalues)
```

```python
final_data={}
for j in range (1,5):
  print('If   %s  principal components(eigenvectors) is/are used'%(j))
  selected_eigenvectors=eigenvectors[:, :j]
  row_zero_mean_data= np.transpose(x_train)
  print('original dataset size = ',row_zero_mean_data.shape) #original data= (150 x 4), transposed= (4 X 150)

  row_feature_vector=np.transpose(selected_eigenvectors)
  print('shape of eigenvector matrix row-wise =',row_feature_vector.shape)#eigen vectors rowwise

  final_data[j] = np.transpose(row_feature_vector @ row_zero_mean_data)
  print('final data shape =',final_data[j].shape) # final data plotted only in the top j principal components discarding other dimensions
  print()
```

Plotting covariance matrix of new data

```python
cov_matrix_new={}
for data in range(1,5):
  new_matrix = np.cov(final_data[data], rowvar=False)
  cov_matrix_new[data]=new_matrix
  # Print the shape of the covariance matrix
  print(new_matrix)
  print("Covariance matrix shape:", new_matrix.shape)
```

##Reconstruction of original data from transformed data

```python
# Perform inverse transform and reconstruct the data
def inverse_transform(transformed_data, eigenvectors, mean):#mean of  original data column-wise
    # Step 1: Multiply the transformed data by the transpose of the eigenvectors
    # transformed data shape = (150,k) where k=1,2,3,4

    inverse_transformed_data = np.dot(transformed_data, eigenvectors.T)

    # Step 2: Add the mean vector to the result
    reconstructed_data = inverse_transformed_data + mean

    return reconstructed_data
```

```python
reconstructed_data={}
for no_of_pcs in range(1,5):
  print('WHEN RECONSTRUCTED FROM %s  PRINCIPAL COMPONENTS(EIGENVECTORS) '%(no_of_pcs))
  print()
  reconstructed_data[no_of_pcs]=(inverse_transform(np.array(final_data[no_of_pcs]),eigenvectors[:, :no_of_pcs],mean))
  print(inverse_transform(np.array(final_data[no_of_pcs]),eigenvectors[:, :no_of_pcs],mean))
  print()
```

##plot reconstruction mean squared loss for different number of components used

```python
#calculating the reconstruction loss
from sklearn.metrics import mean_squared_error
mse_values=[]

for no_of_pcs_used in range(1,5):
  mse = mean_squared_error(X, reconstructed_data[no_of_pcs_used])
  mse_values.append(mse)

print(mse_values)
```

```python
no_of_components=[1,2,3,4]
plt.plot(no_of_components, mse_values, marker='o')
plt.xlabel('Number of Principal Components')
plt.ylabel('Mean Squared Error (MSE)')
plt.text(0.8,0.95, extra_legend, ha='left', va='center', color='red', transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', edgecolor='red', boxstyle='round'))

plt.title('Reconstruction Quality: MSE vs. Number of Principal Components')
plt.xticks(no_of_components)
plt.grid(True)
plt.show()
```
